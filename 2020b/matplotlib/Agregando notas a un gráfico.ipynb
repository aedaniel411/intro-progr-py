# Agregando notas a un gráfico

```python
# importamos modulos
import matplotlib.pyplot as plt

# Definimos gráfico
plt.figure(10)
plt.text(0.25,0.5,'Hola mundo!',fontsize=30)

# Mostramos gráfico
plt.show()
```

## Texto matemático

```python
plt.figure(11)
plt.title(r'$\alpha > \beta$')

plt.show()
```

```python
# Definimos datos
x = range(0,10)
y = [t**2 for t in x]
z = [t**2+1 for t in x]

# Configuramos grafico
plt.figure(50)

plt.plot(x, y, label = r'$\beta=\alpha^2$')
plt.plot(x, z, label = r'$\beta=\alpha^2+1$')

# titulo de ejes
plt.xlabel(r'$\alpha$')
plt.ylabel(r'$\beta$')

plt.legend()

# Mostramos gráfico
plt.show()
```

## Flechas

```python
plt.figure(12)
plt.axis([0, 9, 0, 18])

arrstyles = ['-', '->', '-[', '<-', '<->', 'fancy', 'simple', 'wedge']
for i, style in enumerate(arrstyles):
    plt.annotate(style, xytext=(1, 2+2*i), xy=(4, 1+2*i), arrowprops=dict(arrowstyle=style))

connstyles=["arc", "arc,angleA=10,armA=30,rad=15", "arc3,rad=.2", "arc3,rad=-.2", 
            "angle", "angle3"]
for i, style in enumerate(connstyles):
    plt.annotate("", xytext=(6, 2+2*i), xy=(8, 1+2*i), arrowprops=dict(arrowstyle='->', 
                                                                       connectionstyle=style))

plt.show()
```

```python
import numpy as np

# Definimos datos
nums = np.arange(0,10,0.1)

# Definimos Gráfica
plt.figure (30)
plt.plot(nums, np.sin(nums))
plt.annotate("look!", xy=(np.pi/2, 1), xytext=(5,0), 
             arrowprops=dict(arrowstyle='fancy', 
                             connectionstyle='arc3,rad=0.5'))

# mostramos grafica
plt.show()
```

```python
# Definimos figura
plt.figure(31)
plt.plot(nums, np.sin(nums))
plt.annotate("look!", xy=(np.pi/2, 1), xytext=(5,0), 
             arrowprops=dict(arrowstyle='fancy,head_length=2,head_width=2'))

# Mostramos figura
plt.show()
```

```python
# Definimos figura
plt.figure(32)
plt.plot(nums, np.sin(nums))
plt.annotate("look!", xy=(np.pi/2, 1), xytext=(5,0), arrowprops=dict(arrowstyle='->'))

# Mostramos figura
plt.show()
```

```python
# importamos modulos
import matplotlib.pyplot as plt
import numpy as np

# definomos datos
y = np.random.rand(10)

#definomos gráfico
plt.figure(20)
plt.plot(y) 

# Definimos el rango de valores para el eje y
plt.ylim(-0.2, 1.2) 

# Colocamos texto cerca del valor donde se encuentra el mínimo 
plt.text(np.argmin(y), np.min(y) - 0.1, 'Mínimo', fontsize = 10,
         horizontalalignment='center', verticalalignment='center')
# Colocamos texto cerca del valor donde se encuentra el máximo 
plt.text(np.argmax(y), np.max(y) + 0.1, 'Máximo', fontsize = 10,
         horizontalalignment='center', verticalalignment='center')

# Mostramos gráfico
plt.show()
```

```python
# definomos datos
y = np.random.rand(10)

#definomos gráfico
plt.figure(21)
plt.plot(y)

# Definimos el rango de valores para el eje y
plt.ylim(-0.5, 1.5) 

# Colocamos texto cerca del valor donde se encuentra el mínimo 
plt.text(np.argmin(y), np.min(y) - 0.4, 'Mínimo', fontsize = 10,
         horizontalalignment='center', verticalalignment='center')
# Colocamos texto cerca del valor donde se encuentra el máximo 
plt.text(np.argmax(y), np.max(y) + 0.4, 'Máximo', fontsize = 10,
         horizontalalignment='center', verticalalignment='center')

# Flecha indicadora 
plt.arrow(np.argmax(y), np.max(y) + 0.3, 0, -0.3, 
          length_includes_head = "True", shape = "full", 
          width=0.07, head_width=0.1)
# Flecha indicadora 
plt.arrow(np.argmin(y), np.min(y) - 0.3, 0, 0.3, 
          length_includes_head = "True", shape = "full",
          width=0.07, head_width=0.1)

# Mostramos gráfico
plt.show()
```
